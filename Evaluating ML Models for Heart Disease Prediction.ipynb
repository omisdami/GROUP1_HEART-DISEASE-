```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier


# Assuming 'heart' DataFrame is already loaded
X, y = heart.drop('target', axis=1), heart['target']

# Extract feature names from the DataFrame
feature_names = X.columns

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Train the Random Forest model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Evaluate the model on the test set
accuracy = model.score(X_test, y_test)

print("Model accuracy:", accuracy)


# Get feature importances
importances = model.feature_importances_

# Create a DataFrame for feature importances
feature_importance_df = pd.DataFrame({
    "feature_names": feature_names,
    "importances": importances
}).sort_values(by="importances", ascending=False)

# Print the feature importance rankings
print("feature importance rankings：")
print(feature_importance_df)
```

```python
# List of features to remove
features_to_remove = ['slope', 'sex', 'exang', 'restecg', 'fbs']

# Remove the specified features
X_reduced = X.drop(columns=features_to_remove)

# Split the reduced dataset into training and testing sets
X_train_reduced, X_test_reduced, y_train, y_test = train_test_split(
    X_reduced, y, test_size=0.3, random_state=42, stratify=y
)

# Train the Random Forest model on the reduced dataset
model_reduced = RandomForestClassifier(n_estimators=100, random_state=42)
model_reduced.fit(X_train_reduced, y_train)

# Evaluate the model on the test set
accuracy = model_reduced.score(X_test_reduced, y_test)

print("Model accuracy after removing low-importance features:", accuracy)

# Get the new feature importances
importances_reduced = model_reduced.feature_importances_

# Create a new DataFrame for the reduced feature importances
feature_importance_reduced_df = pd.DataFrame({
    "feature_names": X_reduced.columns,
    "importances": importances_reduced
}).sort_values(by="importances", ascending=False)

# Print the updated feature importance rankings
print("\nUpdated feature importance rankings after removal:")
print(feature_importance_reduced_df)
```
